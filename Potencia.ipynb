```
def powItera(x,y):    
  r=1                   #O(1)
  for i in range(y):    #O(n)
    r*=x                #O(1)
  return r              #O(1)
```

```

def powRec(x, y):
  if y==0:                        #O(n) if else
    return 1                      #O(1)
  else:
    return (powRec(x,(y-1))) *x   
```

```
def powRecDiv(x,n):
  if n==0:
    return 1                          #O(1)
  if (n & 1) != 0:                    #O(log n)
    return x * powRecDiv(x*x, n>>1)
  else:
    return powRecDiv(x*x, n>>1)
```
